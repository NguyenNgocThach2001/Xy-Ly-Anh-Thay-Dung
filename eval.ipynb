```python
from ultralytics import YOLO
from pathlib import Path
import matplotlib.pyplot as plt

#MODEL = "colab_runs/weights/best.pt"
MODEL = "runs/detect/model_on_dataset_combined_500_epoch_medium/weights/best.pt"
DATA  = "Dataset_Detection/piece/Dataset_combined/data.yaml"

model   = YOLO(MODEL)
metrics = model.val(data=DATA, plots=True, save_json=True)

print(f"mAP@0.50      : {float(metrics.box.map50):.4f}")
print(f"mAP@0.75      : {float(metrics.box.map75):.4f}")
print(f"mAP@0.5:0.95  : {float(metrics.box.map):.4f}")
print("Lưu đồ thị tại:", metrics.save_dir)

for name in ["PR_curve.png","F1_curve.png","P_curve.png","R_curve.png","confusion_matrix.png","results.png"]:
    p = Path(metrics.save_dir) / name
    if p.exists():
        img = plt.imread(p); plt.figure(figsize=(6,4))
        plt.imshow(img); plt.axis("off"); plt.title(name); plt.show()
```
